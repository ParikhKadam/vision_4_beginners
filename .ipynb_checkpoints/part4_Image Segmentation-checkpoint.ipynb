```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap = None):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = cmap)
```

## Image Segmentation

## Contour Detection 

```python
coins = cv2.imread('images/pennies.jpg')
coins_gray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)
display(coins_gray, cmap = 'gray')
```

```python
coins_gray = cv2.medianBlur(coins_gray, 25)
_, threshold = cv2.threshold(coins_gray, 200, 255, cv2.THRESH_BINARY_INV)
display(threshold, cmap = 'gray')
```

```python
# Find the contour of the figure
image, contours, hierarchy = cv2.findContours(threshold, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

# For every entry in contours
for i in range(len(contours)):
    
    # last column in the array is -1 if an external contour (no contours inside of it)
    if hierarchy[0][i][3] == -1:
        
        # We can now draw the external contours from the list of contours
        cv2.drawContours(coins, contours, i, (255, 0, 0), 10)
        
display(coins)
```

## WaterShed Algorithm

```python
coins = cv2.imread('images/pennies.jpg')
coins_gray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)
coins_gray = cv2.medianBlur(coins_gray, 25)
```

```python
_, thresh = cv2.threshold(coins_gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
display(thresh, cmap = 'gray')
```

```python
# Remove noise by opening
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = 2)

# sure background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)

display(sure_bg, cmap='gray')
```

```python
# Detect the center of the coins 
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)

thresh_1 = dist_transform.max() * .7
_, sure_fg = cv2.threshold(dist_transform, thresh_1, 255, cv2.THRESH_BINARY)

display(sure_fg, cmap = 'gray')
```
